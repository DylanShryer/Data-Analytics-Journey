What is SQL
    Language that retrieves/transofrms data stored in tables
        A table has columns (variables) and rows (records)
        SQL queries return a result table
        SQL is declarative, so say what you want to get to it
    For analytics, it's mostly selecting data, filtering data, summarizing data, and joining tables

Creating Tables (example)
    Semicolons after each line
    Tables act somewhat like functions
    Utilized in Execute SQL tab

    DROP TABLE IF EXISTS users; #if there is a current user table, delete it so a new one can be created
    DROP TABLE IF EXISTS orders; #if there is a current order table, delete it so a new one can be created

    CREATE TABLE users ( #create user table
    user_id INTEGER PRIMARY KEY, #primary key indentifier for user table
    signup_date TEXT, #attributes
    country TEXT
    );

    CREATE TABLE orders ( 
    order_id INTEGER PRIMARY KEY, #primary key indentifier for user table
    user_id INTEGER, #attributes
    order_date TEXT,
    revenue REAL,
    status TEXT
    );

SELECT
    Choosing columns (my "output")
    Controls what columns appear in results

    Example 1: Shows everything from users
    SELECT *
    FROM users;

    Example 2: Shows user_id and country from users (specific columns)
    SELECT user_id, country
    FROM users;

WHERE
    Filters rows (before any summarizing)
    Decides which rows are included

    Equality Filter
        Example 1: Shows users from the US by equalling an attribute to something
        SELECT *
        FROM users
        WHERE country = 'US';

    Numeric Filter
        Example 1: Shows numerical outputs, like orders where the revenue is over 50
        SELECT *
        FROM orders
        WHERE revenue >= 50;

    Multiple Conditions
        Example 1: Shows orders with a completed status and a revenue over 30 (multiple restrictions at once)
        SELECT *
        FROM orders
        WHERE status = 'completed'
            AND revenue >= 30;

    OR
        Example 1: Shows users either from the US or UK (this or that scenario so another thing can be added)
        SELECT *
        FROM users
        WHERE country = 'US'
            OR country = 'UK';

    IN
        Clean than many ORs

        Example 1: Shows users either from the US or UK
        SELECT *
        FROM users
        WHERE country IN ("US", "UK");


ORDER BY
    Sorts output
    DESC = descending
    ASC = ascending (default)

Highest revenue first
    Example A: Ordered revenue descending
    SELECT *
    FROM orders
    ORDER By revenue DESC;

LIMIT
    Restricts number of rows returned
    Extremely common in things like top 10 products or 50 most recent events

    Example 1: Shows 2 orders from highest to lowest revenue (top 2)
    SELECT *
    FROM orders
    ORDER BY revenue DESC
    LIMIT 2;

Aliases (AS)
    Renames columns in results

    Example 1: Shows column now named user_country instead of just country
    SELECT country AS user_country
    FROM users;

    Example 2: Shows the number of total orders (only the amount, no other info) and a renamed orders title
    SELECT COUNT(*) AS total_orders
        FROM orders;

NULL
    Missing/unknown/unapplicable
    Not equal to anything, even NULL
        IS NULL / IS NOT NULL

    Example 1: Shows no orders, as everything listed has revenue
    SELECT *
    FROM orders
    WHERE revenue IS NULL;

    Example 2: Shows all orders, as everything listed has revenue
    SELECT *
    FROM orders
    WHERE revenue IS NOT NULL;

SQL Style Conventions
  Keywords uppercase (SELECT, FRPOM, WHERE)
  One clause per line
  Indent conditions

  Example 1: Readable SQL
  SELECT
    order_id,
    user_id,
    revenue
  FROM orders
  WHERE status = 'completed'
  ORDER BY revenue DESC;

Mini-Lab
    Figure out what each code example does

Example 1: Select all information (attributes including user_id, signup_date, and country) from users
    SELECT * FROM users;

Example 2: Select user_id and signup_date from users from the country CA
    SELECT user_id, signup_date
    FROM users
    WHERE country = 'CA';

Example 3: Select order_id, user_id, revenue, and status from orders where status is completed, ordering by revenue highest to lowest
    SELECT order_id, user_id, revenue, status
    FROM orders
    WHERE status = 'completed'
    ORDER BY revenue DESC;

Example 4: Select orders, limiting to 2 based on the most recent order dates
    SELECT *
    FROM orders
    ORDER BY order_date DESC
    LIMIT 2;

Problem 2: WHERE filters

    Answer

    b. Show orders with revenue strictly greater than 30

    Answer
    SELECT *
    FROM orders
    WHERE revenue > 30;

    c. Show order_id, status, and revenue of orders that are either completed or that has a revenue >= 50

    Answer
    SELECT order_id, status, revenue
    FROM orders
    WHERE revenue >= 50
        OR status = 'completed';
