Core Formulas and Logical Functions

What Excel formulas are
    Starts with =
    Takes inputs (cells, ranges)
    Produces an output
    Automatically recalculates
    Let you
        Compute metrics
        Apply conditions
        Handle missing values
        Replicate SQL-like logic in spreadsheets
    Row-wise calculations + aggregations

Basic Arithmetic Formulas
    Addition
        =A2 + B2
    Subtraction
        =A2 - B2
    Multiplication
        =A2 * B2
    Division
        =A2 / B2
    Watch for division by zero
        Use IFERROR when appropriate
            =IFERROR(A2 / B2, 0)

Absolute vs Relative References
    Relative Reference
        =A2 * B2
        Changes when copied
    Absolute Reference
        =$A$2 * B2
        Never changes
        Used for
            Tax rates
            constants
            Thresholds
            Lookup tables

Core Aggregation Functions
    SUM
        =SUM(C2:C10)
    AVERAGE
        =AVERAGE(C2:C10)
    MIN/MAX
        =MIN(C2:C10)
        =MAX(C2:C10)
    COUNT Variants
        =COUNT
            Numbers only
            Does not equal the number of rows
        =COUNTA
            Non-empty cells
        =COUNTBLANK
            Blank cells

Conditional Logic
    Basic IF
        =IF(A2 >= 50, "High", "Low")
    Nested IF (use sparingly)
        =IF(A2 >= 80, "Very High", IF(A2>= 50, "High", "Low"))

Logical Operators
    Comparisons
        =
        <>
        >
        <
        >=
        <=
    AND
        =AND(A2 >= 50, B2 = "US")
            Does not evaluate row by row, so if you are trying to iterate through it will just see one FALSE value as FALSE for all
                COUNTIF is the best option for evaluating rows, not just values
        =OR(A2 >= 50, B2 = "US")
            Returns TRUE if at least one value is TRUE from what is highlighted
                Does not evaluate row by row, so if you are trying to iterate through it will just see one TRUE value as TRUE for all
                    COUNTIF is the best option here for evaluating rows, not just values
        =NOT(A2 = "US")
        Combined with IF
            =IF(AND(A2>=50, B2="US"), "Target", "Other")

Conditional Aggregation
    SUMIF
        Finds the sum rows of numbers, with each row being counted if certain criteria are met
        =SUMIF(B:B, "US", C:C)
    COUNTIF
        Counts the rows, with each row being counted if certain criteria are met
        =COUNTIF(C:C, ">=45", C:C)
    AVERAGEIF
        Finds the average of a column, with each row being counted if certain criteria are met
        =AVERAGEIF(C:C, "USA)
    Multiple Conditions
        =SUMIFS
        =COUNTIFS
        =AVERAGEIFS
        Example
            =SUMIFS(C:C, B:B, "US", D:D, "Completed")

    
        

Handling Missing Data Safely
    IFERROR
        Replaces errors with whatever is given
        =IFERROR(A2 / B2, 0)
    ISBLANK
        If there are blanks, then replace it with something
    Change values
        =IF(A2="", 0, A2)

Text Formulas
    TRIM
        Removes extra spaces
    UPPER
        =UPPER(A2)
        Puts everything in uppercase
    LOWER
        =LOWER(A2)
        Puts everything in lowercase
    Concatenation
        Puts things together in somewhat of a phrase

Text-to-Columns and Flash Fill
    Real datasets often contain combined fields like
        "Last, First"
    Text-to-Columns
        Splits a signle column into multiple columns
            1. Select the column
            2. Data -> Text to Columns
            3. Delimited -> Next
            4. Choose Delimiter: Comma
            5. Finish
    Flash Fill
        Learns pattern and fills it automatically
            1. Type the first desired output manually
            2. Start typing the second, so Excel suggests a pattern
            3. Press Enter or Data -> Flash Fill

Common Formula Mistakes
    Forgetting absolute references
    Dividing by zero
    Using COUNT instead of COUNTA
    Hardcoding values instead of referencing cells
    Over-nesting IF instead of using helper columns (why)
