Aggregation
    Raw data is usually too detailed
        Don't want to look at 10,000 orders but rather
            Total revenue
            Average order value
            Orders per user
            Revenue by country
        Turns many rows into summaries

Aggregate Functions
    Take multiple rows and return one value per group
    COUNT(*)
        Number of rows

        Example 1: Counts number of orders
        SELECT COUNT(*) AS total_orders
        FROM orders;
    COUNT(column)
        Number of non-NULL values

        Example 1: Counts orders, including multiple orders (duplicates) from users (like if user 1 appears twice)
        SELECT COUNT(user_id) AS count_user_id
        FROM orders;
    COUNT(DISTINCT column)
        Number of unique values

        Example 1: Counts orders, not including multiple orders (duplicates) from users (like if user 1 appears twice)
        SELECT COUNT(DISTINCT user_id) AS count_user_id
        FROM orders;
    SUM(column)
        Total

        Example 1: Total revenue from orders
        SELECT SUM(revenue) AS total_revenue
        FROM orders;
    AVG(column)
        Average

        Example 1: Average order revenue
        SELECT AVG(revenue) AS avg_order_value
        FROM orders;
    MIN(column)
        Smallest value

        Example 1: Smallest order revenue value
        SELECT MIN(revenue) as smallest_order
        FROM orders;
    MAX(column)
        Largest value
        
        Example 1: Largest order revenue value
        SELECT MAX(revenue) as largest_order
        FROM orders;


Aggregation without GROUP BY
    Grain of result, where 1 row = entire dataset
    Example 1: One row of order results
    SELECT
        COUNT(*) AS orders,
        SUM(revenue) AS revenue,
        AVG(revenue) AS avg_revenue
    FROM orders;

Aggregation with GROUP BY
    Changes what one output row represents
    Every column in SELECT must be grouped or aggregated, so results in SQL are not ambiguous
    Grouping multiple columns create finer grain summaries
    
    Example 1: Multiple rows, showing orders from each user_id (one row per user)
    SELECT
        user_id,
        COUNT(*) AS orders
    FROM orders
    GROUP BY user_id;

    Example 2 (INVALID): A user can have multiple revenues, and SQL doesn't know which one to pick
    SELECT user_id, revenue
    FROM orders
    GROUP BY user_id;

    Example 3: Each user (with a user_id) has a status and order count
    SELECT
        user_id,
        status,
        COUNT(*) AS orders
    FROM orders
    GROUP BY user_id, status;

HAVING to Filter After Aggregation
    Grouped results can not be filtered using where
        WHERE
            rows
        HAVING
            Grouped results
    
    Example 1a (INVALID): Doesn't work, because constraints on multiple things grouped together (user_id, COUNT(*)) can't be constrained before grouping
    SELECT user_id, COUNT(*) AS orders
    FROM orders
    WHERE COUNT(*) >= 2
    GROUP BY user_id;

    Example 1b: Users that have at least 2 orders, using HAVING so that multiple things grouped together can be constrained 
    SELECT
        user_id,
        COUNT(*) AS orders
    FROM orders
    GROUP BY user_id
    HAVING COUNT(*) >= 2;

WHERE and GROUP BY Together
    WHERE filters can be utilized before grouping, as long as the attribute from WHERE isn't a part of the GROUP BY aggregation

    Example 1: Users that have completed orders (can be constrained using WHERE, as a different attribute status was utilized)
    SELECT
        user_id,
        COUNT(*) AS completed_orders
    FROM orders
    WHERE status = 'completed'
    GROUP BY user_id;

Multiple Aggregates Together
    Multiple metrics can be computed at once

    Example 1: User order count, revenue sum, and average revenue from order
    SELECT
        user_id,
        COUNT(*) AS orders,
        SUM(revenue) AS revenue,
        AVG(revenue) AS avg_order_value
    FROM orders
    GROUP BY user_id;

Common Aggregation Mistakes
    Forgetting GROUP BY

    Example 1a (INVALID): user_id is not grouped, so code won't execute
    SELECT user_id, COUNT(*)
    FROM orders;

    Example 1b: User's ID and order count
    SELECT user_id, COUNT(*)
    FROM orders
    GROUP BY user_id;

    Wrong grain assumption

    Example 2a (INVALID): Counts user's orders, not users
    SELECT COUNT(user_id)
    FROM orders;

    Example 2b: Counts total users, not separating them by each order
    SELECT COUNT(DISTINCT user_id)
    FROM orders;

Homework (lesson 2)

Problem 1: Basic Aggregation
    a. What is the average revenue of completed orders only?
    Answer
    SELECT AVG(revenue) AS avg_completed_revenue
    FROM orders
    WHERE status = 'completed';

    b. What is the total revenue across all orders?
    SELECT SUM(revenue) AS total_revenue
    FROM orders;

Problem 2: GROUP BY
    a. For each user_id, return total number of orders, total revenue, and highest single order revenue.
    
    Answer
    SELECT
        user_id,
        COUNT(*) AS order_count,
        SUM(revenue) AS total_revenue,
        MAX(revenue) AS max_order_revenue
    FROM orders
    GROUP by user_id;

Problem 3: Filtering Before Grouping
    a. For each user, how many orders did they place with revenue over or equal to 30?

    Answer
    SELECT
        user_id,
        COUNT(*) AS order_count
    FROM orders
    WHERE revenue >= 30
    GROUP by user_id;

Problem 4: HAVING
    a. Which users have an average order revenue greater than 40?

    Answer
    SELECT
        user_id,
        AVG(revenue) as avg_revenue
    FROM orders
    GROUP by user_id
    HAVING AVG(revenue)>= 40;

Problem 5: COUNT(DISTINCT)
    a. How many distinct users have placed a completed order?

    ANSWER
    SELECT COUNT(DISTINCT user_id) AS buyers_completed
    FROM orders
    WHERE status = 'completed';

Problem 6: Conceptual (written)
    a. What's the grain of Problem 2's result table?
    
    ANSWER
    The grain of the result table is 1 row per user_id, as that is how many users placed orders. Due to the fact that GROUP BY was implemented through user_id, this means that each user will be a row. If GROUP BY didn't occur, the grain would be 1 row total.

    b. Why is that grain different from the grain of the orders table?

    ANSWER
    The grain for the orders table would be 1 row per order, as that each order would be counted as a row. Since one user had 2 orders, that means it only has 3 rows instead of 4 when GROUP BY is utilized under user_id. If this didn't occur, it would also be 4 rows.
