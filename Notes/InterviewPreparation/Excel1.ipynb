Rows, columns, and cells
    Row
        One recond (entity)
    Column
        One attribute (variable)
    Cell
        Single value (row + column)
    Example
        user_id	country	revenue
        1	    US	    50
        2	    CA	    30
        
        Grain
            One row = one user

Data Types in Excel
    Main types
        Text
        Number
        Date
        Boolean (TRUE/FALSE)
    Common Mistakes
        Numbers stored as text
        Dates treated as text
        Mixed types in one column
    How to check
        Select column -> Home -> Number Format
        =ISTEXT(), =ISNUMBER()

Excel Tables
    Always convert raw data into a table
    Steps
        1. Select all data (CTRL + A)
        2. Press CMD + T or navigate to Insert -> Table
        3. Confirm "My table has headers"
    Why this matters
        Formulas auto-fill
        Structured references
        Pivot tables work correctly
        Filters apply cleanly

Sorting and Filtering
    Sorting
        Sort by numeric, date, or text
        Be careful with partial selections (tables prevent this)
    Filtering
        Not delete, but rather temporarily hides rows
        Example
            Filter country = "US"

    Always

        Sort ascending to descending

Conditional Formatting
    Visually flags patterns and problems such as    
        Duplicates
        Values above/below thresholds
        Missing values
        Outliers
    Example
        Highlight duplicates in user_id
            1. Select user_id column
            2. Home -> Conditional Formatting -> Highlight Cell Rules -> Duplicate Values
            3. Choose format -> OK

Data Validation
    Prevents bad data from being entered
        Restrict to list values
        Restrict numeric ranges
        Enforce date ranges
    Example
        Revenue must be non-negative
            1. Select revenue cells
            2. Data Validation
            3. Allow: Decimal
            4. Data: greater than or equal to
            5. Minimum: 0
