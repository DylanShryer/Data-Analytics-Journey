Sorting, Filtering, and Lookup Functions (XLOOKUP / VLOOKUP)

What lookups are
    Lookup
        Matches a value in one table
        Retrieves a related value from another table
    Mental model
        One column defines who you're looking for
        Another column defines what you want to retrieve

Sorting and Filtering
    Sorting safely
        Always sort entire tables
        Tables prevent misalignment
        Example
            Sort revenue descending
            Sort by multiple columns (Count -> Revenue)
    Filtering
        Filters do not delete data
        Filters hid rows temporarily
        Formulas ignore hidden vs visible rows unless explicity told otherwise

Why Lookup Functions Exist
    Imagine
        Orders table has product_id
        Product table has product_name
        You want product_name in the orders table
        This is not copying, this is joining

XLOOKUP
    Syntax
        =XLOOKUP(lookup_value, lookup_array, return_array, "Error handling")
            lookup_value
                What am I trying to match
            lookup_array
                Where should Excel look to find that value
            return_array
                What do I want back when a match is found
            "if not found"
                What should Excel return if there's no match (optional)
    One to one ratio or multiple to one ratio
        There can be one or more lookup keys (one to one or multiple to one), but there can't be multiple associated values (one to multiple or multiple to multiple)
            Will collapse into a single match, usually using the first associated value instead of the others
    Example (does work)
        Find the country of the user who ordered a product (countries will be the same for the same user, so it works)
    Example (doesn't work)
        Find the revenue of the user who ordered a product (revenue may not be the same for the same user if they ordered multiple things, so it doesn't work)

VLOOKUP
    Syntax
        =VLOOKUP(lookup_value, table_array, col_index, range_lookup)
            lookup_value
                What am I trying to match
            table_array
                Entire lookup table (lookup column must be first)
            col_index_num
                Which column should be returned
            range_lookup
                FALSE
                    Exact match
                TRUE
                    Approximate match
    Problems
        Column order matters
        Can break if columns move
        Only looks right, not left
        One to one ratio or multiple to one ratio

FILTER
    Syntax
        =FILTER(array, condition)
            For same table
        AND condition
            =FILTER(array, (condition) * (condition))
        OR condition
            =FILTER(array, (condition) + (condition))
        TRUE
            1
        FALSE
            0
    There can be one of more lookup keys (one to multiple or multiple to multiple) associated with different values
    Example
        Completed orders
            =FILTER(orders, orders[status] = "completed")
FILTER + MATCH
    Is this key in a list of allowed keys
    Syntax
        =FILTER(array, ISNUMBER(MATCH(array_of_keys, list_of_allowed_keys, 0)))
            FILTER()
                array
                    Rows you want back
            MATCH()
                Exact match
                    array_of_keys
                        Key column from those rows
                    list_of_allowed_keys
                        Keys that should be kept
            ISNUMBER()
                Converts matches into TRUE/FALSE
        Example
            Only revenues from CA users
